```python
data = data.drop_duplicates()
```

```python
missing_values = data.isnull().sum()
missing_values
```

```python
data = data.drop(['Height(CM)', 'Weight(KG)', 'hobbies', 'social medai & video'], axis=1)
data = data.loc[:, data.apply(pd.Series.nunique) != 1]
data.info()
```

```python
data.head(30)
```

```python
data['college mark group'] = pd.cut(data['college mark'], bins=[0, 50, 60, 70, 80, 90], labels=['F', 'D', 'C', 'B', 'A'])
```

```python
data['college mark group'].value_counts().plot.pie(autopct='%1.1f%%')
```

```python
# Basic data visualization
sns.histplot(data['college mark'])
```

```python
data['salary group'] = pd.cut(data['salary expectation'], bins=[0, 50000, 75000, 100000], labels=['low salary','median salary','high salary'])
```

```python
data['salary group'].value_counts().plot.bar()
```

```python
gender_group = data.groupby('Gender')['college mark'].mean().rename('Avg_CollegeMark_By_Gender')
gender_group
```

```python
financial_group = data.groupby('Financial Status')['college mark'].mean().rename('Avg College Mark by Financial Status')
financial_group
```

```python
job_group = data.groupby('part-time job')['college mark'].mean().rename('Avg College Mark by Part-time Job')
job_group
```

```python
data.rename(columns={'daily studing time': 'daily studying time'}, inplace=True)
data.columns
```

```python
studying_time = data.groupby('daily studying time')['college mark'].mean().rename('Avg College Mark by Daily Studying Time')
studying_time
```

```python
data['daily studying time'].value_counts().plot.bar()
```
